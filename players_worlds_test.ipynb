```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
matches = pd.read_csv('worlds_matches.csv')
averages = pd.read_csv('player_avg_df.csv')
```

```python
matches.replace('TSM', 'Team SoloMid', inplace = True)
averages.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
averages_red = averages.copy()
columns_blue = averages.columns
```

```python
columns_red = []
for column in averages_red.columns:
    columns_red.append('opp_' + column)
    
averages_red.columns = columns_red
averages_red
```

```python
cl = averages.columns.tolist() + averages_red.columns.tolist() + ['result']
```

```python
full_matches = []
for i in range(len(matches)):
    blue_team = averages[averages['team'] == matches.iloc[i]['blue_team']].values.tolist()
    red_team = averages[averages['team'] == matches.iloc[i]['red_team']].values.tolist()
    
    temp = blue_team[0] + red_team[0] + [matches.iloc[i]['result']]
    full_matches.append(temp)
```

```python
df = pd.DataFrame(full_matches, columns = cl)
df
```

```python
df.to_csv('players_worlds_test.csv')
```
