```python
columns_red = []
for column in columns_blue:
    columns_red.append("opp_" + column)
    
#columns_red
```

```python
cl = all_avg.columns.tolist() + averages_red.columns.tolist() + ['result']
#cl
```

```python
teams_names_df = pd.read_csv('team_df_names.csv')
#teams_names_df
```

```python
matches = teams_names_df[['team', 'opp_team', 'result']]
#matches
```

```python
full_matches = []
for i in range(len(matches)):
    blue_team = all_avg[all_avg['team'] == matches.iloc[i]['team']].values.tolist()
    red_team = all_avg[all_avg['team'] == matches.iloc[i]['opp_team']].values.tolist()
    
    temp = blue_team[0] + red_team[0] + [matches.iloc[i]['result']]
    full_matches.append(temp)
```

```python
df = pd.DataFrame(full_matches, columns = cl)
```
