#### Se haran varios cambios para poder realizar la tarea de prediccion que se tiene en mente
- se combinara las entradas de ambos equipos a una unica entrada que contiene la informacion de ambos equipos
- el atributo result cambiara de ser representado por 1 el vencedor a 1 = victoria equipo azul, -1 = victoria equipo rojo
- todas las columnas que contengan informacion "espejo" es decir representen informacion global del juego solo se contendran una vez

#### cambie las preferencias de pandas para poder ver la informacion completa de cada entrada, esto es para determinar de que manera se hara la condensacion de informacion, debemos recordar que cada 2 entradas equivale a un unico juego, por lo tanto debemos considerar como unir ambas entradas para poder llevar a cabo la tarea de prediccion que se tiene planeada para etapas posteriores

#### Limpiando aun mas nuestro dataframe de equipos para condensar toda la informacion en una unica entrada. El dataframe nos facilita un poco la tarea gracias a la existencia de todas las entradas con el prefijo "opp_" todas estas entradas son la informacion del oponente, asi que unicamente se debe de determinar que informacion del segundo euqipo no se encuentra ya en la entrada original del primer equipo y agregarla con un prefijo "opp_"
