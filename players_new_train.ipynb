```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
avg_dataframe = pd.read_csv('player_avg_df.csv')
test_matches = pd.read_csv('player_avg_test.csv')
```

```python
avg_dataframe.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
matches = test_matches[['team', 'opp_team', 'result']]
#matches
```

```python
averages_red = avg_dataframe.copy()
```

```python
columns_blue = avg_dataframe.columns
#columns_blue
```

```python
columns_red = []
for column in averages_red.columns:
    columns_red.append('opp_' + column)
    
averages_red.columns = columns_red
averages_red
```

```python
cl = avg_dataframe.columns.tolist() + averages_red.columns.tolist() + ['result']
```

```python
full_matches = []
for i in range(len(matches)):
    blue_team = avg_dataframe[avg_dataframe['team'] == matches.iloc[i]['team']].values.tolist()
    red_team = avg_dataframe[avg_dataframe['team'] == matches.iloc[i]['opp_team']].values.tolist()
    
    temp = blue_team[0] + red_team[0] + [matches.iloc[i]['result']]
    full_matches.append(temp)
```

```python
df = pd.DataFrame(full_matches, columns = cl)
df
```

```python
df.to_csv('players_train_df.csv')
```
