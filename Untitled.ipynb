### Lista de atributos utilizados
A continuación se encuentran descritos los atributos utilizados pare realizar la agrupación de estilos de juego por los equipos profesionales de League of Legends que participaron en el campeonato mundial de League of Legends en 2020.

Los atributos utilizados para la descripción de los estilos de juego que representan los equipos provienen de dos distintas fuentes:

- El panel oficial de análisis de Riot Games 
- De mi propia autoría

Todos los atributos que no sean mencionados de mi propia autoría provienen del panel de análisis oficial de Riot Games

### Atributos para describir equipos
- gamelenght:
- - Describe cual fue la duración promedio de las partidas jugadas por el equipo (valores menores son mejores)
- winrate:
- - Describe el porcentaje de victorias que los equipos acumularon con respecto a todas las partidas que jugaron
- KDA
- - Atributo compuesto, el cual es calculado a partir de la suma de los asesinatos y las asistencias y dividido entre las muertes de un jugador o un equipo
- firstblood:
- - Porcentaje de veces que un equipo logro obtener el primer asesinato del juego
- firstdragon:
- - Porcentaje de veces que un equipo logro obtener el primer dragón del juego
- dragons:
- - Promedio de dragones derrotados por partida de un equipo
- firstherald:
- - Porcentaje de veces que un equipo logro obtener el primer heraldo del juego
- heralds:
- - Promedio de heraldos derrotados por partida de un equipo
- barons:
- - Promedio de barones derrotados por partida de un equipo
- firsttower:
- - Porcentaje de veces que un equipo destruyo la primer torreta enemiga del juego
- towers:
- - Promedio de torretas destruidas por juego
- dpm:
- - Daño por minuto promedio de un equipo
- vspm:
- - Puntaje de visión por minuto de un equipo promediado por cada partida
- visionscore:
- - Promedio del puntaje de visión por partida de un equipo
- earned gpm:
- - Promedio de oro generado por minuto por partida

### Atributos para describir el aprovechamiento de ventajas de un equipo
- FirstBlood%:
- - Porcentaje de primera sangre obtenida en el total de partidas jugadas
- FbGoldAt10% (First Blood Gold At 10 %):
- - Atributo de mí autoría
- - Describe el porcentaje de partidas donde un equipo logro conseguir el primer asesinato del juego y consiguió mantener una ventaja de oro desde ese punto hasta el minuto 10
- FbFirstDragon% (First Blood First Dragon %):
- - Describe el porcentaje de partidas en las que un equipo logro obtener el primer asesinato del juego y ademas el primer dragón del juego
- FbFirstHerald% (First Blood First Herald %):
- - Describe el porcentaje de partidas en las que un equipo logro obtener el primer asesinato del juego y ademas el primer heraldo del juego
- FbFirstTower% (First Blood First Tower %):
- - Describe el porcentaje de partidas en las que un equipo logro obtener el primer asesinato del juego y logro destruir la primer torreta del juego
- GoldAt10%:
- - Describe el porcentaje de partidas en las que un equipo logro obtener una ventaja de oro al minuto 10
- Gd10GoldAt15% (Gold at 10 Gold at 15 %):
- - Atributo de mí autoría
- - Describe el porcentaje de veces que un equipo logro mantener su ventaja de oro al minuto 10 y llevarla a una ventaja al minuto 15 del juego
- GoldAt15%:
- - Describe el porcentaje de partidas en las que un equipo logro obtener una ventaja al minuto 15 del juego
- Gd15FirstTo3Towers%:
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un equipo obtuvo una ventaja al minuto 15 y ademas logro destruir 3 torretas primero
- Gd15FirstMidTower%:
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un equipo obtuvo una ventaja al minuto 15 y ademas logro destruir la primer torreta del carril intermedio
- FirstTo3Towers%:
- - Describe el porcentaje de partidas en las que un equipo logro ser el primero en destruir 3 torretas enemigas
- Ft3FirstBaron% (First to 3 Towers First Baron %)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un equipo fue el primero en destruir tres torretas enemigas y logro derrotar el primer barón del juego

### Atributos para describir a los tres jugadores de un carril principal (linea superior, intermedia e inferior), nos permite encontrar la afinidad que tiene el equipo con este jugador (que tantos recursos ponen sobre dicho jugador) y que tan bien puede este mismo jugador mantener o incrementar las ventajas.

- roleFB%: (role: bot/mid/top)
- - Describe el porcentaje de veces que ese jugador en específico obtuvo el primer asesinato de la partida
- roleFbFirstHerald%: (role: bot/mid/top)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un jugador en específico obtuvo el primer asesinato de la partida y posteriormente el equipo obtuvo el primer heraldo del juego
- roleFbFirstDragon%: (role: bot/mid/top)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un jugador en específico obtuvo el primer asesinato de la partida y posteriormente el equipo obtuvo el primer dragón del juego
- roleFbGoldAt10%: (role: bot/mid/top)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un jugador en específico obtuvo el primer asesinato de la partida y el equipo logro obtener la ventaja de oro al minuto 10
- fbGoldAt10Role%: (role: Bot/Mid/Top)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un jugador en específico obtuvo el primer asesinato de la partida y el jugador logro mantener una ventaja de oro en comparación a su oponente de línea
- roleFbCsDriff%: (role: bot/mid/top)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un jugador en específico obtuvo el primer asesinato de la partida y el jugador logro conseguir una ventaja en el número de súbditos al minuto 10
- roleFbXpDriff%: (role: bot/mid/top)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un jugador en específico obtuvo el primer asesinato de la partida y el jugador logro conseguir una ventaja en experiencia del personaje al minuto 10 (experiencia es usada para subir de nivel)
- roleFbFirstTower%: (role: bot/mid/top)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un jugador en específico obtuvo el primer asesinato de la partida y ademas el equipo logro destruir la primer torreta enemiga del juego
- roleGd10%: (role: bot/mid/top)
- - Describe el porcentaje de partidas en las que un jugador obtuvo una ventaja de oro al minuto 10 del juego en comparación a su oponente de línea
- roleGd10Gd15%: (role: bot/mid/top)
- - Describe el porcentaje de partidas en las que un jugador obtuvo una ventaja de oro al minuto 10 del juego en comparación a su oponente de línea y logro mantener la ventaja hasta el minuto 15 del juego
- roleGd15%: (role: bot/mid/top)
- - Describe el porcentaje de partidas en las que un jugador obtuvo una ventaja de oro al minuto 15 del juego en comparación a su oponente de línea
- roleGd15Wr%: (role: bot/mid/top)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un jugador obtuvo una ventaja de oro al minuto 15 del juego en comparación a su oponente de línea y ademas su equipo logro ganar la partida
- roleGd15Firstto3%: (role: bot/mid/top)
- - Atributo de mí autoría
- - Describe el porcentaje de partidas en las que un jugador obtuvo una ventaja de oro al minuto 15 del juego en comparación a su oponente de línea y ademas su equipo logro destruir primero 3 torretas enemigas
- roleXp10%: (role: bot/mid/top)
- - Describe el porcentaje de partidas en las que un jugador logro obtener una ventaja de experiencia en comparación a su oponente de línea al minuto 10
- roleXp10Xp15%: (role: bot/mid/top)
- - Describe el porcentaje de partidas en las que un jugador logro obtener una ventaja de experiencia en comparación a su oponente de línea al minuto 10 y ademas logro mantener la ventaja de experiencia hasta el minuto 15
- roleCs10%: (role: bot/mid/top)
- - Describe el porcentaje de partidas en las que un jugador logro obtener una ventaja de súbditos al minuto 10 en comparación a su oponente de línea
- roleCs10Cs15%: (role: bot/mid/top)
- - Describe el porcentaje de partidas en las que un jugador logro obtener una ventaja de súbditos al minuto 10 en comparación a su oponente de línea y ademas logro mantener la ventaja de súbditos hasta el minuto 15

### Atributos para describir al jugador del rol "jungla", muchos de los atributos ya han sido descritos anteriormente así que unicamente describiré aquellos que aún no han sido descritos

- enemyMonstersPerGame:
- - Describe cuantos monstruos de la sección del oponente puede robar un jugador por partida
- roleWardsKilledPerGame: (roles: sup/jng)
- - Describe cuantos centinelas de visión destruye por partida un jugador
- roleWardsPerMinute: (roles: sup/jng)
- - Describe cuantos centinelas de visión coloca por minuto por partida un jugador

### Atributos para describir al jugador del rol "soporte" muchos de los atributos ya han sido descritos anteriormente así que unicamente describiré aquellos que aún no han sido descritos

- supDamageShare:
- - Describe que tanto daño hizo el jugador del rol "soporte" en comparación al resto del equipo
- supDamageMitigatedPerGame:
- - Describe cuanto daño recibió y mitigo el jugador del rol "soporte"
- supWardsPlacedPerGame:
- - Describe en promedio cuantos centinelas de visión coloca el jugador del rol "soporte" por juego
- supAvgWardsPerMinute
- - Describe la cantidad de centinelas de visión por minuto por juego pone el jugador del rol "soporte"
- supCWardsBought:
- - Describe la cantidad de centinelas de control que compra por juego el jugador del rol "soporte"
- supVSPerGame:
- - Describe el puntaje de visión por juego que obtuvo el jugador del rol "soporte"
- supAvgVSPerMinute:
- - Describe el puntaje de visión por minuto que obtuvo el jugador del rol "soporte"
