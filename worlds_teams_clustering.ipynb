```python
team_names = graph_atributes['team']

#team_names
```

```python
graph_atributes.drop('team', axis = 1, inplace = True)
```

```python
normalized_graph_atributes=(graph_atributes-graph_atributes.min())/(graph_atributes.max()-graph_atributes.min())
normalized_graph_atributes
```

```python
categories = normalized_graph_atributes.columns.tolist()
categories = [*categories, categories[0]]
```

```python
team1 = normalized_graph_atributes.iloc[80].values.tolist()
team2 = normalized_graph_atributes.iloc[205].values.tolist()
```

```python
team1 = [*team1, team1[0]]
team2 = [*team2, team2[0]]
```

```python
label_loc = np.linspace(start = 0, stop = 2 * np.pi, num=len(team1))
```

```python
plt.figure(figsize = (8,8))
plt.subplot(polar = True)
plt.plot(label_loc, team1, label = 'DAMWON Gaming')
plt.plot(label_loc, team2, label = 'Suning')
plt.title('team comparison', size = 20, y = 1.05)
lines, labels = plt.thetagrids(np.degrees(label_loc), labels = categories)
plt.legend
plt.show
```

```python
from sklearn.cluster import KMeans
```

```python
OMP_NUM_THREADS=1
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters = i, init = 'random', n_init = 10, max_iter = 500, tol = 1e-04, random_state = 43)
    km.fit(normalized_graph_atributes)
    distortions.append(km.inertia_)
```

```python
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5).fit(normalized_graph_atributes)
```

```python
labels = pd.DataFrame(kmeans.labels_)

labeledTeams = pd.concat((normalized_graph_atributes, labels), axis = 1)
labeledTeams
```

```python
labeledTeams
```

```python
labeledTeams['constant'] = 'Data'
```

```python
labeledTeams = labeledTeams.rename({0 : 'labels'}, axis = 1)
```

```python
f, axes = plt.subplots(4, 4, figsize = (20, 15), sharex = False)
f.subplots_adjust(hspace = 0.2, wspace = 0.7)

for i in range(0, len(list(labeledTeams)) - 2):
    col = labeledTeams.columns[i]
    if i < 4:
        ax = sns.stripplot(x = labeledTeams['constant'], y = labeledTeams[col].values, 
                           hue = labeledTeams['labels'], jitter = True, ax = axes[0, (i)])
        ax.set_title(col)
    
    elif i >= 4 and i < 8:
        ax = sns.stripplot(x = labeledTeams['constant'], y = labeledTeams[col].values, 
                           hue = labeledTeams['labels'], jitter = True, ax = axes[1, (i-4)])
        ax.set_title(col)
        
    elif i >= 8 and i < 12:
        ax = sns.stripplot(x = labeledTeams['constant'], y = labeledTeams[col].values, 
                           hue = labeledTeams['labels'], jitter = True, ax = axes[2, (i-8)])
        ax.set_title(col)
        
    elif i >= 12:
        ax = sns.stripplot(x = labeledTeams['constant'], y = labeledTeams[col].values, 
                           hue = labeledTeams['labels'], jitter = True, ax = axes[3, (i-12)])
        ax.set_title(col)
```
